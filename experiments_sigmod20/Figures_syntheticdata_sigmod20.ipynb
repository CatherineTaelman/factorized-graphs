# Figures appearing in the Algebraic Amplification paper at SIGMOD2020
This notebook allows one to generate every synthetic data figure appearing in the Algebraic Amplification paper at SIGMOD 2020. Becacuse certain figures are time and compute intensive, each figure is generated from cached data. 

Each type of experiment has its own Python module (e.g. Fig_Timing, Fig_EndtoEnd_accuracy, etc.) imported at the top of this notebook. These modules define long lists of alternative parameters used for experiments. We choose sets of parameters with metaparameters CHOICE (which choice of parameters are chosen) and VARIANT (which variant for printing results is chosen). Other parameters define control variables like whether new data is created and whether figures are displayed or exported as PDF.
